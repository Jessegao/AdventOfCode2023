```python
import re
# Question 3
lines = text.split('\n')
total = 0

def gamePossible(game):
    if game:
        id = re.search('Game (.*):', game).group(1)
        game = game.split(': ')[1]
        for round in game.split('; '):
            for cube in round.split(', '):
                # print(cube.split(' '))
                num, color = cube.split(' ')
                num = int(num)
                if not ((color == 'red' and num <= 12) or \
                    (color == 'green' and num <= 13) or \
                    (color == 'blue' and num <= 14)):
                    return 0
        return id
    return 0

for l in lines:
    total += int(gamePossible(l))
print(total)
```

```python
import re
# Question 4
lines = text.split('\n')
total = 0

def gamePossible(game):
    if game:
        id = re.search('Game (.*):', game).group(1)
        game = game.split(': ')[1]
        max_dice = {
            'red':0,
            'green':0,
            'blue':0,
        }
        for round in game.split('; '):
            for cube in round.split(', '):
                # print(cube.split(' '))
                num, color = cube.split(' ')
                num = int(num)
                max_dice[color] = max(max_dice[color], num)
        return max_dice['blue']*max_dice['green']*max_dice['red']
    return 0

for l in lines:
    total += int(gamePossible(l))
print(total)
```
